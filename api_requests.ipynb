```python
import time

import spotifyRequests as sp
import pandas as pd
```

```python
spotify = sp.SpotifyRequests()
```

# Preparation the uris

```python
# Load the dataset
df = pd.read_csv('datasets/full_streaming_2017to2023.csv')
df_year = []

# Separate for year
for year in range(2017, 2024):
   df_year.append(df[(df['ts'] >= f'{year}-01-01') & (df['ts'] <= f'{year}-12-31')]) 

print(df_year)
```

```python
# Save the datasets in years
for dataframe in df_year:
    year = pd.to_datetime(dataframe['ts'].min())
    dataframe.to_csv(f'Streaming_History_{year.year}.csv')
```
